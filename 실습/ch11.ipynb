```python
import pandas as pd
from scipy.stats import pearsonr
from scipy.stats import spearmanr
from scipy.stats import kendalltau
```

```python
df = pd.read_csv("bike.csv")
df.head(2)
```

```python
df.corr()
```

```python
df[["casual", "registered", "count"]].corr()
```

```python
stat,p = pearsonr(df["casual"], df["registered"])
```

```python
print(stat) # 각각의 상관계수
print(p) # p-value
```

```python
# 계절별로 체감온도와 자전거 대여 숫자의 상관관계를 알아보고자 한다.
# 이 때 상관계수는?
```

```python
df[["season", "atemp", "casual"]].groupby("season").corr()
```

```python
df_corr = df[["season", "atemp", "casual"]].groupby("season").corr()
df_corr = df_corr.reset_index()
df_corr
```

```python
df_corr = df_corr.loc[df_corr["atemp"] < 1, ]
df_corr
```

```python
# 날씨에 따른 기온과 자전거 대여의 상관계수 변화를 알아보고자 한다
# 날씨가 맑은날 (weather = 1)과 그렇지 않은 날의 상관계수 차이의 절대값은?
```

```python
df.head(2)
```

```python
df["is_sunny"] =  (df["weather"] == 1) + 0
df.head(2)
```

```python
df_corr = df[["is_sunny", "temp", "casual"]].groupby("is_sunny").corr()
df_corr
```
