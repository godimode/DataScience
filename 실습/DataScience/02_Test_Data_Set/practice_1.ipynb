```python
import pandas as pd
```

```python
df = pd.read_csv("TV.csv")
df.head(2)
```

```python
df_sub = df[["Frequency", "Picture_quality", "Speaker"]]
df_sub.head(2)
```

```python
df_sub["x1"] = df_sub["Frequency"].str.contains("60 Hz") + 0
df_sub["x2"] = df_sub["Picture_quality"].str.contains("60 Hz") + 0
df_sub["x3"] = df_sub["Speaker"].str.contains("60 Hz") + 0
df_sub.head(2)
```

```python
df_sub["x4"] = df_sub.loc[:,"x1":"x3"].sum(axis = 1)
df_sub.head(2)
```

```python
df_sub["x4"].value_counts()
```

```python
df_sub2 = df[["Stars", "Operating_system", "channel", "Picture_quality"]]
df_sub2.head(2)
```

```python
df_sub2["4K"] = df_sub2["Operating_system"].str.contains("4K") + df_sub2["channel"].str.contains("4K") + df_sub2["Picture_quality"].str.contains("4K") + 0
df_sub2["8K"] = df_sub2["Operating_system"].str.contains("8K") + df_sub2["channel"].str.contains("8K") + df_sub2["Picture_quality"].str.contains("8K") + 0
df_sub2.head(5)
```

```python
df_sub2.head(10)
```

```python
df_sub2["4K_result"] = df_sub2["4K"] * df_sub2["Stars"]
df_sub2["8K_result"] = df_sub2["8K"] * df_sub2["Stars"]
df_sub2.head(10)
```

```python
df_sub3 = df.loc[~df["channel"].str.contains("Oper|Pixel"), :]
df_sub3.head(10)
```

```python
df_sub3["review_rate"] = df_sub3["Reviews"] / df_sub3["Ratings"]
df_sub3["discount"] = df_sub3["current_price"] / df_sub3["MRP"]
df_sub3.head(2)
```

```python
df_sub3["netflix"] = df_sub3["channel"].str.contains("Netflix") + 0
df_sub3["prime"] = df_sub3["channel"].str.contains("Prime Video") + 0
df_sub3["4k8k"] = df_sub3["Picture_quality"].str.contains("4K|8K") + 0
df_sub3.head(2)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
df_sub3 = df_sub3[["Stars", "review_rate", "MRP", "discount", "netflix", "prime", "4k8k"]].copy()
df_sub3.head(2)
```

```python
model = RandomForestRegressor(random_state=123)
model.fit(y = df_sub3["Stars"], X = df_sub3.iloc[:, 1:])
model.feature_importances_
```
