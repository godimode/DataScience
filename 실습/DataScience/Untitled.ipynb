```python
data = pd.read_csv("data/bike.csv")
```

```python
data.head(2)
```

```python
data["datetime"] = pd.to_datetime(bike["datetime"])
data["date"] = data["datetime"].dt.date
data.head(2)
```

```python
a = data.groupby(["date"])["humidity"].max()
a
```

```python
data_h100 = data.groupby(["date"])["humidity"].max().reset_index()
data_h100.head(2)
```

```python
data_h100 = data_h100.loc[data_h100["humidity"] == 100]
data_h100.head(2)
```

```python
len(data_h100)
```

```python
data_join = pd.merge(left=data, right=data_h100, 
                    left_on="date", right_on="date", 
                    how="inner")
data_join.head(2)
```

```python
data_join_temp_up30 = data_join.loc[data_join["temp"] >= 30]
data_join_temp_up30.head(2)
```

```python
data_join_temp_up30["count"].mean()
```

```python
data_join_temp_up30["count"].all()
```

```python
data_join_temp_up30["count"].sum()/len(data_join_temp_up30)
```
