```python
df_8_test2 = pd.read_csv("bike.csv")
df_8_test2.head(2)
```

```python
df_8_test2["season"].value_counts()
```

```python
df_8_test2 = df_8_test2.loc[df_8_test2["season"] % 2 == 0, ]
df_8_test2.head(2)
```

```python
df_8_summer = df_8_test2.loc[df_8_test2["season"] == 2, ["datetime", "registered"]]
df_8_winter = df_8_test2.loc[df_8_test2["season"] == 4, ["datetime", "registered"]]
```

```python
df_8_summer["hour"] = pd.to_datetime(df_8_summer["datetime"]).dt.hour
df_8_winter["hour"] = pd.to_datetime(df_8_winter["datetime"]).dt.hour
```

```python
df_8_summer.head(2)
```

```python
df_8_summer = df_8_summer.drop(columns="datetime")
df_8_summer.head(2)
```

```python
df_8_winter = df_8_winter.drop(columns="datetime")
df_8_winter.head(2)
```

```python
df_8_summer_sub = df_8_summer.groupby("hour")["registered"].mean().reset_index()
df_8_winter_sub = df_8_winter.groupby("hour")["registered"].mean().reset_index()
```

```python
df_8_ans = pd.merge(left = df_8_summer_sub, right = df_8_winter_sub, left_on = "hour", right_on="hour")
df_8_ans.head(2)
```

```python
df_8_ans["ans"] = abs(df_8_ans["registered_x"] - df_8_ans["registered_y"])
df_8_ans["ans"].idxmax()
```

```python
df_8_test3 = pd.read_csv("bike.csv")
df_8_test3.head(2)
```

```python
df_8_test3["date"] = pd.to_datetime(df_8_test3["datetime"])
df_8_test3.head(2)
```

```python
df_8_test3["dd"] = df_8_test3["date"].dt.date
df_8_test3.head(2)
```

```python
df_8_test3["hour"] = df_8_test3["date"].dt.hour
df_8_test3.head(2)
```

```python
df_8_test3_rain = df_8_test3.groupby("dd")["humidity"].max().reset_index()
df_8_test3_rain.head(2)
```

```python
df_8_test3_rain = df_8_test3_rain.loc[df_8_test3_rain["humidity"] == 100, ]
df_8_test3_rain.head()
```

```python
len(df_8_test3_rain)
```

```python
df_8_test3_ans = pd.merge(left=df_8_test3_rain, right = df_8_test3.loc[:, ["dd", "temp", "count", "humidity", "hour"]], left_on= "dd", right_on="dd")
```

```python
df_8_test3_ans.head(2)
```

```python
df_8_test3_ans = df_8_test3_ans.loc[df_8_test3_ans["temp"] > 30, ]
df_8_test3_ans.head(2)
```

```python
df_8_test3_ans.groupby("hour")["count"].mean().reset_index().mean()
```

### CH09 데이터 전처리 : 정렬 및 변환

```python
dia = pd.read_csv("diamonds.csv")
dia.head(2)
```

```python
pd.crosstab(dia["cut"], dia["clarity"])
```

```python
pd.crosstab(dia["cut"], dia["clarity"], normalize=True)
```

```python
pd.crosstab(dia["cut"], dia["clarity"], normalize=True).round(2)
```

```python
pd.crosstab(dia["cut"], dia["clarity"], values=dia["price"], aggfunc=pd.Series.mean)
```

```python
dia.groupby(["cut", "clarity"])["price"].mean().reset_index()
```

```python
dia_agg = dia.groupby(["cut", "clarity"])["price"].mean().reset_index()
dia_agg.head(2)
```

```python
dia_agg.sort_values("price").head()
```

```python
dia_agg.sort_values("price", ascending=False).head()
```

```python
dia_agg.sort_values(["cut", "clarity"], ascending=[True, False]).head()
```
